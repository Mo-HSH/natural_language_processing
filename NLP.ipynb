
# Natural Language Processing

```python
# Download necessary packages
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = "Racism remains a pervasive issue in society, manifesting through various forms of discrimination that affect individuals based on their background. Promoting diversity is essential for fostering an environment of justice, where everyone is treated as equal and has access to the same rights. It is crucial to challenge prejudice and cultivate tolerance among communities to achieve true equality for all."
text
```

**Segmentation**

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)
sentences
```

**Tokenization**

```python
from nltk.tokenize import word_tokenize

words = word_tokenize(text)
print(words)
```

**Remove Stop Words**

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]
print(filtered_words)
```

**Stem and Lemma**

```python
!pip install matplotlib
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed = [stemmer.stem(word) for word in filtered_words]
lemmatized = [lemmatizer.lemmatize(word) for word in filtered_words]

print("Stemmed Words:", stemmed)
print("Lemmatized Words:", lemmatized)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Convert lists to strings
stemmed_text = ' '.join(stemmed)
lemmatized_text = ' '.join(lemmatized)

# Create WordClouds
stemmed_wordcloud = WordCloud(width=400, height=200, background_color='white').generate(stemmed_text)
lemmatized_wordcloud = WordCloud(width=400, height=200, background_color='white').generate(lemmatized_text)

# Set up the matplotlib figure
plt.figure(figsize=(10, 5))

# Plot Stemmed Word Cloud
plt.subplot(1, 2, 1)
plt.imshow(stemmed_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Stemmed Words')

# Plot Lemmatized Word Cloud
plt.subplot(1, 2, 2)
plt.imshow(lemmatized_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Lemmatized Words')

# Show the plots
plt.tight_layout()
plt.show()
```
